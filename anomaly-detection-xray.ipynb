```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(640,640),
        batch_size=32,
        class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,  
    target_size=(640, 640),
    batch_size=32,
    class_mode='categorical',
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size = (640, 640),
    batch_size= 32,
    class_mode = 'categorical',
    )
```

```python
import tensorflow as tf

from tensorflow.keras import models, Input,layers
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam, SGD, RMSprop

import matplotlib.pyplot as plt

import numpy as np
import pandas as pd

import random
```

```python


model_ADXR = models.Sequential(name = 'Anomaly-Detection_XRay')
model_ADXR.add(Input(shape=(640,640,3)))

model_ADXR.add(Conv2D(32, (3, 3), activation='relu',strides=1 , name='conv1'))
model_ADXR.add(MaxPooling2D((2, 2), name='pooling1'))

model_ADXR.add(Conv2D(64, (3, 3), activation='relu', name='conv2'))
model_ADXR.add(MaxPooling2D((2, 2), name='pooling2'))

model_ADXR.add(Conv2D(128, (3, 3), activation='relu', name='conv3'))
model_ADXR.add(MaxPooling2D((2, 2), name='pooling3'))

model_ADXR.add(Conv2D(256, (3, 3), activation='relu', name='conv4'))
model_ADXR.add(MaxPooling2D((2, 2), name='pooling4'))

model_ADXR.add(Flatten())

model_ADXR.add(Dense(128, activation='relu', name='dense1'))
model_ADXR.add(Dense(50,activation='relu'))

model_ADXR.add(Dense(4, activation='softmax', name='sortie'))

model_ADXR.compile(optimizer='adam',
                  loss='categorical_crossentropy',  
                  metrics=['accuracy'])



model_ADXR.summary()
```

```python
model_ADXR.compile(optimizer = 'adam', loss='categorical_crossentropy', metrics = ['accuracy'])
```

```python
%%time 
# Essayez de chronometrer les modèles lors de l'apprentissage
#verbose, epochs, batch_size, steps_per_epoch = 1, 5, 32, 20

from tensorflow.keras.callbacks import EarlyStopping

# Définir le callback EarlyStopping
early_stopping = EarlyStopping(
    monitor='val_loss',  # Surveillez la perte de validation
    patience=3,  # Nombre d'époques sans amélioration avant d'arrêter
    verbose=1,  # Affiche les messages lorsqu'il y a un arrêt anticipé
    restore_best_weights=True  # Restaure les poids du modèle au meilleur moment
)

history = model_ADXR.fit(
    train_generator,
    steps_per_epoch=10,
    epochs=20,
    validation_data=val_generator,
    validation_steps=16,
    callbacks=[early_stopping]
)
```

```python
_ , accuracy = model_ADXR.evaluate(test_generator)
print(f'accuracy : {accuracy}')
```
